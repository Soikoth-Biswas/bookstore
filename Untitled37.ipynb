```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Generate some data points
x = np.linspace(0, 10, 100)
y = np.sin(x)

# 2. Create the plot
plt.figure(figsize=(8, 4))
plt.plot(x, y, label="Sine Wave", color="purple", linewidth=2)

# 3. Add titles and labels
plt.title("My First Google Colab Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid(True)
plt.legend()

# 4. Display the graph
plt.show()
```
